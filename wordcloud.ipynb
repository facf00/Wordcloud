```python
class CubicWordCloud(WordCloud):
    
    def __init__(self, width=400, height=200, background_color='white', min_font_size=10, font_path=None):
        super().__init__(width=width, height=height, background_color=background_color,
                         min_font_size=min_font_size, font_path=font_path)
        self.interpolation = 'cubic'
    
    def _make_fit(self, frequency):
        # compute the cubic polynomial coefficients
        x = np.arange(len(frequency))
        coeffs = np.polyfit(x, frequency, 3)
        # evaluate the polynomial at non-grid points
        xp = np.linspace(0, len(frequency) - 1, num=1000)
        yp = np.polyval(coeffs, xp)
        return np.vstack((xp, yp)).T


# create CubicWordCloud object
wordcloud = CubicWordCloud(width=1600, height=627, background_color='black', min_font_size=1, font_path='Roboto-Regular.ttf')
# generate the word cloud
wordcloud.generate_from_text(' '.join(words_wild))

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
print(len(words))
```
