## A very simple wild wordcloud

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
#!pip install wordcloud
```

```python
words_wild = pd.read_csv('words_wild.csv')
words_full = pd.read_csv('words_full.csv')
```

```python
# create WordCloud object
wordcloud = WordCloud(width=1600,
                      height=627,
                      background_color='white',
                      min_font_size=5).generate(' '.join(words_wild))

# plot the wordcloud
plt.figure(figsize=(8, 8), dpi=300)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()

# sum of all words
print(words_wild.shape)
```

```python
# save to png
wordcloud.to_file("world_cloud_wild_navy.png");
```

```python
# create CubicWordCloud object
wordcloud = WordCloud(width=1600,
                      height=627,
                      background_color='black',
                      min_font_size=10,
                      font_path='Roboto-Regular.ttf')
# generate the word cloud
wordcloud.generate_from_text(' '.join(words_wild))

plt.figure(figsize=(10, 10), dpi=300)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
print(words_wild.shape)
```

```python
# save to png
wordcloud.to_file("world_cloud_wild.png");
```

```python
# create WordCloud object
wordcloud = WordCloud(width=1600,
                      height=627,
                      background_color='white',
                      min_font_size=1,max_font_size=50,
                      font_path='Roboto-Regular.ttf')

# generate the word cloud
wordcloud.generate_from_text(' '.join(words_full))

# plot the word cloud
plt.figure(figsize=(10, 10), dpi=300)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
print(words_full.shape)
```
